### Store Data (25 Marks)

#### Storing cleaned and formatted dataframes into an SQL database

```python
# sqlite3 has already been imported in the beginning of this coursework.
```

```python
# Storing the dataframes into the database with exception handling.
try:
    sqliteConnect = sqlite3.connect('ResultDatabase.db')
    dfFormattedCleanMockTest.to_sql('FormativeMockTest',
                               sqliteConnect,
                               if_exists = 'replace')
    dfFormattedCleanTest1.to_sql('FormativeTest1',
                               sqliteConnect,
                               if_exists = 'replace')
    dfFormattedCleanTest2.to_sql('FormativeTest2',
                               sqliteConnect,
                               if_exists = 'replace')
    dfFormattedCleanTest3.to_sql('FormativeTest3',
                               sqliteConnect,
                               if_exists = 'replace')
    dfFormattedCleanTest4.to_sql('FormativeTest4',
                               sqliteConnect,
                               if_exists = 'replace')
    dfFormattedCleanSumTest.to_sql('SummativeTest',
                               sqliteConnect,
                               if_exists = 'replace')
    print('Success! The dataframes are now stored in the database.')
    sqliteConnect.close()
    
except sqlite3.Error as error:
        print('Failed to store the dataframes into the database', error)
finally:
    if sqliteConnect:
        sqliteConnect.close()
        print('The SQLite connection is closed.')
```
